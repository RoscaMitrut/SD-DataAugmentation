```python
# Importing libraries
import tkinter as tk
from tkinter import Label
from tkinter import filedialog
from PIL import Image, ImageTk
import numpy as np

# Global variable to store the original image
original_img = None

# Image uploader function
def imageUploader():
    global original_img
    fileTypes = [("Image files", "*.png;*.jpg;*.jpeg")]
    path = tk.filedialog.askopenfilename(filetypes=fileTypes)

    # if file is selected
    if len(path):
        original_img = Image.open(path)  # Store the original image
        img = original_img.resize((400, 400))  # Resize image to fit the layout
        pic = ImageTk.PhotoImage(img)

        # Set the original image on the left label
        label.config(image=pic)
        label.image = pic

    else:
        print("No file is chosen! Please choose a file.")

# Function to add random noise to the original image (RGB)
def addNoise():
    global original_img
    if original_img:
        # Resize the original image
        img_resized = original_img.resize((400, 400))

        # Convert the image to a NumPy array
        img_array = np.array(img_resized)

        # Generate random noise
        noise = np.random.randint(0, 50, img_array.shape, dtype='uint8')  # Noise intensity between 0 and 50

        # Add the noise to the original image
        noisy_img_array = img_array + noise

        # Clip values to stay within valid range [0, 255]
        noisy_img_array = np.clip(noisy_img_array, 0, 255)

        # Convert the noisy image back to a PIL Image
        noisy_img = Image.fromarray(noisy_img_array)

        # Convert the noisy image to an ImageTk object for display
        pic_noisy = ImageTk.PhotoImage(noisy_img)

        # Set the noisy image on the right label
        bw_label.config(image=pic_noisy)
        bw_label.image = pic_noisy
    else:
        print("No image loaded to add noise.")

# Main method
if __name__ == "__main__":

    # Defining tkinter object
    app = tk.Tk()

    # Setting title and fixed size of the app
    app.title("Image Viewer with Noise Addition")
    app.geometry("900x550")  # Fixed size
    app.resizable(False, False)  # Make window non-resizable

    # Frame to hold both images side by side
    image_frame = tk.Frame(app)
    image_frame.pack(pady=20)

    # Label to show the original image (left side)
    label = tk.Label(image_frame)
    label.pack(side=tk.LEFT, padx=10)

    # Label to show the noisy image (right side)
    bw_label = tk.Label(image_frame)
    bw_label.pack(side=tk.RIGHT, padx=10)

    # Defining upload button
    uploadButton = tk.Button(app, text="Locate Image", command=imageUploader)
    uploadButton.pack(pady=10)

    # Defining button to add noise to the image
    noiseButton = tk.Button(app, text="Add Noise to Image", command=addNoise)
    noiseButton.pack(pady=10)

    app.mainloop()
```
